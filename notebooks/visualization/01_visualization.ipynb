```python
import pandas as pd

# Update the file path to your actual CSV file

df = pd.read_csv(r'C:\Users\ACER\Documents\Capstone\E-commerce-Price-Analytics\amazon_price_history_load_script\s24_price_in_thb.csv', parse_dates=['date'])
df['date'] = pd.to_datetime(df['date'])
df['price_thb'] = pd.to_numeric(df['price_thb'], errors='coerce')
df = df.dropna()
df = df.sort_values(by='date')
df = df.reset_index(drop=True)
df['moving_average_7d'] = df['price_thb'].rolling(window=7).mean()
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(14, 7))
sns.lineplot(data=df, x='date', y='price_thb', label='Daily Price (THB)')
sns.lineplot(data=df, x='date', y='moving_average_7d', label='7-Day Moving Average', color='orange')
plt.title('Price in THB Over Time with 7-Day Moving Average')
plt.xlabel('Date')
plt.ylabel('Price (THB)')
plt.legend()
plt.grid(True)
plt.show()
```
